## Negative Indexing
Negative indexing means start from the end

-1 refers to the last item, -2 refers to the second last item etc.

```python
print(nums[-1])
print(nums[-2])
print(nums[-3])
len(nums)
```

```python
# starting from ending of the list
for num in range(10,len(nums),-1):   # there start,stop,gap concept is used
    print(nums[num])
```

```python
for num in range(len(nums) - 1, -1, -1):
    print(nums[num])

"""
    Explanation

    len(nums) - 1 → last index of the list

    -1 → stop before -1 (so it includes index 0)

    -1 → move backward
"""
```
