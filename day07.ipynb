```python
# check the number is positive , negative or 0
num = int(input("Enter the number: "))
if num == 0:
    print("Give number is 0 ")
elif num <0:
    print("Number is negative ")
else:
    print("Number is Positive ")
```

```python
# print day according to its number let say monday is in 1 position and so on

day_num = int(input("Enter day number: "))
if day_num == 1:
    print("The day is Monday")
elif day_num == 2:
    print("The day is Tuesday")
elif day_num == 3:
    print("The day is Wednesday")
elif day_num == 4:
    print("The day is Thursday")
elif day_num == 5:
    print("The day is Friday")
elif day_num == 6:
    print("The day is Saturday")
elif day_num == 7:
    print("The day is Sunday")

else:
    print("Invalid number! Please enter number between 1 and 7")
```
